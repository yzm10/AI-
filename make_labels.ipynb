```python
df['gender_label'] = df['gender'].apply()
```

```python
df = df.dropna()
```

```python
def gender2int(label):
    g = {
        'Boys' : 0,
        'Girls': 1,
        'Men': 2,
        'Unisex': 3,
        'Women': 4
    }
    return g[label]

def master2int(label):
    m = {
        'Accessories' : 0,
        'Apparel' : 1,
        'Footwear' : 2,
        'Free Items' : 3,
        'Home' : 4,
        'Personal Care' : 5,
        'Sporting Goods' : 6
    }
    return m[label]

def sport2int(label):
    g = {
        'Casual' : 0,
        'Ethnic' : 1,
        'Formal' : 2,
        'Home' : 3,
        'Party' : 4,
        'Smart Casual' :5,
        'Sports' : 6,
        'Travel' : 7
    }
    return g[label]
```

```python
df['gender_labels'] = df['gender'].apply(gender2int)
df['category_labels'] = df['masterCategory'].apply(master2int)
df['sport_labels'] = df['Sports'].apply(sport2int)
```

```python
df.to_csv('train_change.csv', index=False)
```
